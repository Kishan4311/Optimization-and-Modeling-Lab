### Q1 part 1

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x1, x2):
    return 100 * x2**2 + 100 * x1**4 - 200 * x2 * x1**2 + x2**2 - 2 * x1 + 1


x1 = np.linspace(-2, 2, 500)
x2 = np.linspace(-2, 2, 500)
x1, x2 = np.meshgrid(x1, x2)
z = f1(x1, x2)


plt.figure(figsize=(8, 6))
contour = plt.contour(x1, x2, z, levels=np.linspace(-10, 200, 17), cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8, fmt="%.1f")  
plt.colorbar(contour)


min_x1, min_x2 = 1, 1  
min_value = f1(min_x1, min_x2)  


plt.plot(min_x1, min_x2, 'ro', label=f'Minima at ({min_x1}, {min_x2})')
plt.text(min_x1, min_x2 + 0.2, f"{min_value:.2f}", color='red', fontsize=10, ha='center')  


plt.legend()
plt.title("Level Sets of Function 1")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(alpha=0.3)


plt.show()
```

### Q1 part2

```python
def f2(x1, x2):
    return x1**4 + x2**4 - 20*x1**3 - 20*x2**3 + 100*x1**2 + 100*x2**2


x1 = np.linspace(-4, 4, 500)
x2 = np.linspace(-4, 4, 500)
x1, x2 = np.meshgrid(x1, x2)
z = f2(x1, x2)


min_x1, min_x2 = 2.5, 2.5  
min_value = f2(min_x1, min_x2)


plt.figure(figsize=(8, 6))
contour = plt.contour(x1, x2, z, levels=np.linspace(-500, 5000, 30), cmap='plasma')
plt.clabel(contour, inline=True, fontsize=8, fmt="%.1f")  # Add contour labels
plt.colorbar(contour)





plt.title("Level Sets of Function 2")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(alpha=0.3)


plt.show()
```

### Q1 part 3

```python
def f3(x1, x2):
    return -np.cos(x1) * np.cos(x2) * np.exp(-((x1 - np.pi)**2 + (x2 - np.pi)**2))


x1 = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x2 = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x1, x2 = np.meshgrid(x1, x2)
z = f3(x1, x2)


min_x1, min_x2 = np.pi, np.pi  
min_value = f3(min_x1, min_x2)


plt.figure(figsize=(8, 6))
contour = plt.contour(x1, x2, z, levels=np.linspace(-1, 0, 20), cmap='cividis')
plt.clabel(contour, inline=True, fontsize=8, fmt="%.2f")  
plt.colorbar(contour)


plt.plot(min_x1, min_x2, 'ro', label=f'Minima ({min_x1:.2f}, {min_x2:.2f})\nValue: {min_value:.2f}')
plt.legend()


plt.title("Level Sets of Function 3")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(alpha=0.3)

# Show the plot
plt.show()
```

### Q1 part 4

```python
def f4(x1, x2):
    return 2*x1**2 - 1.05*x1**4 + x1**6/6 + x1*x2 + x2**2


x1 = np.linspace(-2, 2, 500)
x2 = np.linspace(-2, 2, 500)
x1, x2 = np.meshgrid(x1, x2)
z = f4(x1, x2)


min_x1, min_x2 = 0, 0  
min_value = f4(min_x1, min_x2)


plt.figure(figsize=(8, 6))
contour = plt.contour(x1, x2, z, levels=np.linspace(-10, 10, 30), cmap='coolwarm')
plt.clabel(contour, inline=True, fontsize=8, fmt="%.2f")  
plt.colorbar(contour)


plt.plot(min_x1, min_x2, 'ro', label=f'Minima ({min_x1:.2f}, {min_x2:.2f})\nValue: {min_value:.2f}')
plt.legend()


plt.title("Level Sets of Function 4")
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.grid(alpha=0.3)


plt.show()
```
