```python
def avg1(df):
    returns = pd.DataFrame()
    for col in df.columns[1:]:
        s_p = df[col].values
        s_r = []
        for i in range(len(s_p) - 5):
            t_average = (s_p[i + 5] - s_p[i]) / s_p[i]
            s_r.append(t_average)
        returns[col] = s_r
    return returns
t_average = avg1(df)
t_average.mean()
```

```python
from pyomo.environ import *
from pyomo.opt import SolverFactory

# Compute expected returns and covariance matrix using your functions
mu = t_average.mean()  # Expected return of each stock
cov_matrix = covariance_matrix  # Covariance matrix

# Define total investment amount
total_investment = 1_00_00_000  # 1 crore

# Number of stocks
n = len(mu)

# Create Pyomo model
model = ConcreteModel()

# Decision variables: amount invested in each stock
model.x = Var(range(n), domain=NonNegativeReals)

# Objective: Minimize portfolio variance (risk)
model.obj = Objective(expr=sum(
    model.x[i] * cov_matrix.iloc[i, j] * model.x[j]
    for i in range(n) for j in range(n)
), sense=minimize)

# Constraint 1: Total investment constraint
model.total_investment = Constraint(expr=sum(model.x[i] for i in range(n)) == total_investment)

# **Updated Constraint 2: Minimum return constraint with investment factor**
model.min_return = Constraint(expr=sum(mu.iloc[i] * model.x[i] for i in range(n))
                              >= 0.00025 * sum(model.x[i] for i in range(n)))

# Solve the model using Gurobi
solver = SolverFactory('gurobi')
solver.solve(model)

# Display optimized investments
optimized_investments = [model.x[i]() for i in range(n)]
investment_allocation = pd.Series(optimized_investments, index=mu.index)

print("Optimal investment in each stock:")
print(investment_allocation)
```

```python
def optimize_portfolio(covariance_matrix, t_average_mean, target_return, total_investment=1e7):
    model = gp.Model('Portfolio_Optimization')
    model.setParam('OutputFlag', 0)
    n = len(t_average_mean)
    x = model.addVars(n, lb=0.0, name="invest")
    obj = gp.QuadExpr()
    for i in range(n):
        for j in range(n):
            obj.add(x[i] * x[j] * covariance_matrix.iloc[i, j])
    model.setObjective(obj, GRB.MINIMIZE)
    model.addConstr(gp.quicksum(x[i] for i in range(n)) == total_investment)
    model.addConstr(gp.quicksum(x[i] * t_average_mean[i] for i in range(n)) >= target_return * total_investment)
    model.optimize()
    return model.ObjVal, [x[i].X for i in range(n)] if model.status == GRB.OPTIMAL else (None, None)

# Adjusted variable names
t_average_mean = t_average.mean()
return_rates = np.linspace(t_average_mean.min(), t_average_mean.max(), 27)
results = [
    (target_return, *optimize_portfolio(covariance_matrix, t_average_mean, target_return))
    for target_return in return_rates
]

# Extract risk and return values
risk_values, return_values = zip(*[(np.sqrt(risk), r) for r, risk, _ in results if risk is not None])

# Plot Efficient Frontier
plt.plot(risk_values, return_values, 'b-o', linewidth=2)
plt.xlabel('Portfolio Risk (Std)')
plt.ylabel('Expected Return')
plt.title('Risk Vs Return')
plt.grid(True)
plt.show()

# Find minimum risk and maximum return portfolios
min_risk_index = np.argmin(risk_values)
max_return_index = np.argmax(return_values)

# Minimum risk portfolio
min_risk_portfolio = results[min_risk_index]
min_risk_return = min_risk_portfolio[0]
min_risk_value = risk_values[min_risk_index]

# Maximum return portfolio
max_return_portfolio = results[max_return_index]
max_return_value = return_values[max_return_index]

# print(f"Minimum Risk Portfolio:")
# print(f"Target Return: {min_risk_return}")
# print(f"Risk (Std Dev): {min_risk_value}")
# print(f"Investments: {min_risk_portfolio[2]}")

# print(f"\nMaximum Return Portfolio:")
# print(f"Target Return: {max_return_portfolio[0]}")
# print(f"Risk (Std Dev): {risk_values[max_return_index]}")
# print(f"Investments: {max_return_portfolio[2]}")
```

# 1(7)

### **Decision Variables:**
- $ x_i $ : Amount invested in asset \( i \) (where \( i \in \{1,2, \dots, n\} \)).

### **Parameters:**
- $ \mu_i $ : Expected return of asset \( i \).
- $ \Sigma_{ij} $ : Covariance between asset \( i \) and asset \( j \).
- \( R \) : Risk limit (variance constraint).
- \( T \) : Total investment amount.

### **Objective Function:**
$
\max \sum_{i=1}^{n} \mu_i x_i
$
(Maximize the expected return of the portfolio.)

### **Constraints:**
#### **1. Total Investment Constraint:**
$
\sum_{i=1}^{n} x_i = T
$
(The total investment should be exactly equal to \( T \).)

#### **2. Risk Constraint:**
$
\sum_{i=1}^{n} \sum_{j=1}^{n} x_i x_j \Sigma_{ij} \leq R T^2
$
(The total portfolio variance (risk) should be within the allowed risk limit.)

### **Reformulation in Terms of Portfolio Weights \( w_i \):**
Define portfolio weights as:
$
w_i = \frac{x_i}{T}, \quad \sum_{i=1}^{n} w_i = 1
$

Then, the **objective function** becomes:
$
\max T \sum_{i=1}^{n} \mu_i w_i
$

And the **risk constraint** transforms into:
$
\sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \Sigma_{ij} \leq R
$

### **Final Problem Formulation:**
$
\max \sum_{i=1}^{n} \mu_i w_i
$
$
\text{s.t.} \quad \sum_{i=1}^{n} w_i = 1, \quad w_i \geq 0, \quad \sum_{i=1}^{n} \sum_{j=1}^{n} w_i w_j \Sigma_{ij} \leq R
$

- yes its consistent


```python
import warnings

warnings.filterwarnings("ignore")

import gurobipy as gp
from gurobipy import GRB
import matplotlib.pyplot as plt
import numpy as np

def optimize_portfolio_risk_constrained(covariance_matrix, t_average_mean, risk_limit, total_investment=1e7):
    try:
        model = gp.Model('Portfolio_Optimization')
        model.setParam('OutputFlag', 0) 
        
        n = len(t_average_mean)
    
        x = model.addVars(n, lb=0.0, name="investments")
    
        obj = gp.quicksum(x[i] * t_average_mean[i] for i in range(n))
        model.setObjective(obj, GRB.MAXIMIZE)
        model.addConstr(gp.quicksum(x[i] for i in range(n)) == total_investment, "budget")
        risk_expr = gp.QuadExpr()
        for i in range(n):
            for j in range(n):
                risk_expr += x[i] * x[j] * covariance_matrix.iloc[i, j]
        model.addConstr(risk_expr <= risk_limit * total_investment * total_investment, "risk")
        
        model.optimize()
        
        if model.status == GRB.OPTIMAL:
            weights = np.array([x[i].X for i in range(n)]) / total_investment
            optimal_return = model.ObjVal
            actual_risk = np.sqrt(weights.T @ covariance_matrix @ weights)
            
            return optimal_return, weights, actual_risk
        else:
            return None, None, None
            
    except gp.GurobiError as e:
        print(f"Error: {e}")
        return None, None, None

def plot_efficient_frontier_risk_constrained(covariance_matrix, t_average_mean, stock_names):
    risk_limits = np.linspace(0.0001, 0.001, 20)  
    returns = []
    risks = []
    
    for risk_limit in risk_limits:
        result = optimize_portfolio_risk_constrained(covariance_matrix, t_average_mean, risk_limit)
        if result[0] is not None:
            returns.append(result[0])
            risks.append(result[2])
    
    plt.figure(figsize=(10, 6))
    plt.plot(risks, returns, 'b-o', linewidth=2, markersize=4)
    plt.xlabel('Portfolio Risk (Std)')
    plt.ylabel('Expected Return')
    plt.title('Risk-Constrained Optimization')
    plt.grid(True)
    
    plt.show()
    
    return risks, returns

risks, returns = plot_efficient_frontier_risk_constrained(covariance_matrix, t_average_mean, stock_names)

risk_limit = 0.0005 
optimal_return, weights, actual_risk = optimize_portfolio_risk_constrained(covariance_matrix, t_average_mean, risk_limit)


```

```python
def analyze_equal_weight_portfolio(cov_matrix, mean_returns, stock_names, total_investment=1e7):
    n_stocks = len(stock_names)
    equal_weights = np.ones(n_stocks) / n_stocks
    expected_return = np.sum(mean_returns * equal_weights)
    portfolio_risk = np.sqrt(equal_weights.T @ cov_matrix @ equal_weights)
    investment_per_stock = total_investment / n_stocks
    print("\n" + "="*50)
    print("EQUAL-WEIGHT PORTFOLIO ANALYSIS")
    print("="*50)
    
    print("\n1. Portfolio Metrics:")
    print(f"Expected Return: {expected_return:.6f} ({expected_return*100:.3f}%)")
    print(f"Portfolio Risk (σ): {portfolio_risk:.6f} ({portfolio_risk*100:.3f}%)")
    print(f"\nNumber of Stocks: {n_stocks}")
    print(f"Investment per Stock: ₹{investment_per_stock:,.2f}")
    
    print("\n2. Individual Stock Allocations:")
    for stock, weight in zip(stock_names, equal_weights):
        print(f"{stock}: {weight*100:.2f}% (₹{weight*total_investment:,.2f})")
    print("\n3. Comparison with Optimized Portfolios:")
    risk_levels = {
        'Low Risk': 0.0002,
        'Medium Risk': 0.0005,
        'High Risk': 0.0008
    }
    
    print("\nOptimized Portfolio Returns at Different Risk Levels:")
    for risk_name, risk_limit in risk_levels.items():
        optimal_return, _, actual_risk = optimize_portfolio_risk_constrained(
            cov_matrix, mean_returns, risk_limit
        )
        if optimal_return is not None:
            return_difference = optimal_return - expected_return
            print(f"\n{risk_name} (σ = {risk_limit:.4f}):")
            print(f"  Optimized Return: {optimal_return:.6f} ({optimal_return*100:.3f}%)")
            print(f"  Return Difference: {return_difference:.6f} ({return_difference*100:.3f}%)")
            print(f"  Relative Improvement: {(return_difference/expected_return)*100:.1f}%")
    
    plt.figure(figsize=(12, 8))
    risk_limits = np.linspace(0.0001, 0.001, 20)
    returns = []
    risks = []
    
    for risk_limit in risk_limits:
        result = optimize_portfolio_risk_constrained(cov_matrix, mean_returns, risk_limit)
        if result[0] is not None:
            returns.append(result[0])
            risks.append(result[2])
    
    plt.plot(risks, returns, 'b-o', linewidth=2, markersize=4, label='Efficient Frontier')
    plt.plot(portfolio_risk, expected_return, 'r*', markersize=15, label='Equal-Weight Portfolio')
    plt.annotate('Equal-Weight Portfolio', 
                xy=(portfolio_risk, expected_return), 
                xytext=(portfolio_risk+0.0001, expected_return+0.0001),
                arrowprops=dict(facecolor='black', shrink=0.05))
    
    plt.xlabel('Portfolio Risk (Standard Deviation)')
    plt.ylabel('Expected Portfolio Return')
    plt.title('Equal-Weight Portfolio vs Efficient Frontier')
    plt.grid(True)
    plt.legend()
    plt.show()
    return {
        'expected_return': expected_return,
        'portfolio_risk': portfolio_risk,
        'weights': equal_weights,
        'investment_per_stock': investment_per_stock
    }
```
